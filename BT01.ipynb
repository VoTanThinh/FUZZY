```python
import pandas as pd
import numpy as np
url = 'http://Winterolympicsmedals.com/medals.csv'
data = pd.read_csv(url)
data
data[data['NOC']== 'CAN'] #lấy dữ liệu nước canada
data[(data['NOC']=='CAN')&(data['Medal']=='Gold')] # lấy huy chương vàng của tất cả các năm của nước CANADA
data[(data['NOC']=='CAN')&(data['Year']==2006)&(data['Medal']=='Gold')]           #số huy chương vàng nước canada đạt được năm 2006
data[(data['NOC']=='CAN')&(data['Year']==2006)&(data['Medal']=='Silver')]         #số huy chương Bạc nước canada đạt được năm 2006
data[(data['NOC']=='CAN')&(data['Year']==2006)&(data['Medal']=='Bronze')]        #số huy chương Đồng nước canada đạt được năm 2006
data['Medal'].value_counts()                                                     #đếm số huy chương vàng bạc trong cột huy chương của toàn bộ  dữ liệu
data[(data['NOC']=='CAN')&(data['Medal'].count())]
data.groupby((data['NOC']=='CAN')&(data['Year']==2006)&(data['Medal']=='Gold')).count()      #đếm số huy chương vàng năm 2006 của canada
data.groupby((data['NOC']=='CAN')&(data['Year']==2006)&(data['Medal']=='Silver')).count()    #đếm số huy chương bạc năm 2006 của canada
data.groupby((data['NOC']=='CAN')&(data['Year']==2006)&(data['Medal']=='Bronze')).count()    #đếm số huy chương đồng năm 2006 của canada
data.groupby((data['NOC']=='CAN')&(data['Year']==2006)& (data['Medal'])).count()                 # tổng huy chương đạt được năm 2006(vàng+bạc+đồng)
# code lấy dữ liệu 
time = pd.date_range ('16/02/2022',periods = 100000 , freq = '30s')
data = pd.DataFrame (index = time)
data['Acess_Amount']=np.random.randint(1,10,100000)                                           #thời gian lấy mẫu là 30s cứ 1 tuần lấy mẫu 1 lần
data.resample('w').sum()                                                                      # w là tuần , D là này , M là tháng
```
