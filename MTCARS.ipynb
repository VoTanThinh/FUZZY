```python
from importlib import import_module
import pandas as pd
import numpy as np
from sklearn import linear_model
url = 'mtcars.csv'

data = pd.read_csv(url)
correlation = data.corr()
print(correlation)
X = data.drop(data.columns[[0,2]], axis = 1)
data.plot(kind = 'scatter', x= 'wt', y = 'mpg', figsize = (4,9), color = 'black')


```

```python
from importlib import import_module
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
model=linear_model.LinearRegression()
X = pd.DataFrame([data['wt'],data['wt']**1]).T   #CHUYỂN VỊ T a1X^2+a2*X+c

model.fit(X,y=data['mpg']) #DẠY CHO NÓ
y=data['mpg']
print('intercept của hàm bậc 1')
print(model.intercept_)
print('coeffient của hàm bậc 1')
print(model.coef_)
print('độ chính các của hàm mô hình')
print(model.score(X,y))


range=np.arange(1.5,5.5,0.1)
predictor=pd.DataFrame([range,range**1]).T    #bậc 1
y_pred=model.predict(X=predictor)
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range,y_pred,color='blue')


```
